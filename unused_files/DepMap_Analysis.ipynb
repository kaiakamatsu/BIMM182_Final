## Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Collecting the Data

- CRISPR Gene Dependency data extracted from DepMap portal
- https://depmap.org/portal/download/all/

## Exploratory Data Analysis 

```python
# read in file 
df = pd.read_csv("../CRISPRGeneDependency/CRISPRGeneDependency.csv")
```

```python
print(df.columns) # columns = gene
```

```python
print(df['Unnamed: 0']) # row = cell lines
```

```python
RAE1 = df['RAE1 (8480)']
plt.hist(RAE1, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('RAE1 gene dependency scores across cell lines')
plt.ylabel('Frequency')
plt.show()
```

```python
A1BG = df['A1BG (1)']
plt.hist(A1BG, bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('A1BG gene dependency scores across cell lines')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation of Gene Essentiality Across Cell Lines')
plt.show()
```
